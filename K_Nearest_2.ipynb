```python
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
import warnings
from matplotlib import style
from collections import Counter
style.use('fivethirtyeight')
```

```python
dataset = {'cat':[[1,2],[2,3],[3,1]] , 'dog':[[6,5],[7,7],[8,6]], 'rabbit':[[6,2],[7,3],[8,1]]}
color=['green','red','blue','black']
new_features=[5,3]
k=0
for i in dataset:
    for j in dataset[i]:
        plt.scatter(j[0],j[1],s=100,color=color[k])
    k+=1
plt.scatter(new_features[0],new_features[1],s=100, color=color[k])
plt.show()
```

```python
def k_nearest(data,predict,k=3):
    if len(data)>k:
        warnings.warn("idiot")
    distances=[]
    for group in data:
        for features in data[group]:
#             dist=np.sqrt(np.sum((np.array(features)-np.array(predict))**2))
            dist=np.linalg.norm(np.array(features)-np.array(predict))
            distances.append([dist,group])
    votes=[i[1] for i in sorted(distances)[:k]]
    print(Counter(votes).most_common())
    vote_result=Counter(votes).most_common(1)[0][0]
    return vote_result
```

```python
k=len(dataset)
if k%2==0:
    k=k+1
else:
    k=k+2
pred=k_nearest(dataset,new_features,4)
print(pred)
# plt.show()
```
